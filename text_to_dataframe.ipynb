```python
import pandas as pd
import numpy as np
```

```python
file_input = open('chapter19.txt','r')
answers = pd.read_csv('chapter19_answers.txt')
array = np.array(file_input.readlines())
newarray = np.resize(array, (20,6))
df = pd.DataFrame(newarray, columns=['question', 'ans1', 'ans2', 'ans3', 'ans4', 'correct'])
```

```python
answers
```

```python
for i in df.index:
    if df.loc[i,'question'][1] == ".":
        df.loc[i,'question'] = (str(df.loc[i,'question'])[2:-6])+"\n"
    else:
        df.loc[i,'question'] = str(df.loc[i,'question'])[3:-6] + "\n"
    df.loc[i,'correct'] = "ANSWER: " + answers.loc[i][1].strip() + '\n'
#    df.loc[i,'ans1'] = str(df.loc[i,'ans1'])[2:]
#    df.loc[i,'ans2'] = str(df.loc[i,'ans2'])[2:]
#    df.loc[i,'ans3'] = str(df.loc[i,'ans3'])[2:]
#    df.loc[i,'ans4'] = str(df.loc[i,'ans4'])[2:]
```
